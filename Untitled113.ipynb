```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# Configurar el estilo de las gráficas
sns.set(style="whitegrid")



# Intentar cargar el archivo CSV en un DataFrame
try:
  df = pd.read_csv('/content/Museos_05b.csv', encoding='ISO-8859-1', error_bad_lines=False)
  print("Archivo CSV cargado correctamente.")
except pd.errors.ParserError as e:
  print("Error al cargar el archivo CSV:", e)
except FileNotFoundError as e:
  print("Error: No se encontró el archivo en la ruta especificada:", e)
except Exception as e:
  print("Error inesperado:", e)


df = None # Definir df como None en caso de error
# Mostrar las primeras filas del DataFrame solo si df está definido
if df is not None:
  print("Primeras filas del DataFrame:")
  print(df.head())
else:
  print("No se pudo cargar el DataFrame debido a un error.")
# Obtener una descripción estadística del DataFrame solo si df está definido
if df is not None:
  print("\nDescripción estadística del DataFrame:")
  print(df.describe())





```

```python
# Convertir las columnas de fecha a formato de fecha (si existen)
if df is not None and 'Fecha de Ingreso' in df.columns and 'Fecha de Egreso' in df.columns:
  df['Fecha de Ingreso'] = pd.to_datetime(df['Fecha de Ingreso'])
  df['Fecha de Egreso'] = pd.to_datetime(df['Fecha de Egreso'])

  # Calcular la duración de la estancia (si aplicable)

  if df is not None and 'Fecha de Ingreso' in df.columns and 'Fecha de Egreso' in df.columns:
    df['Duración de Estancia (días)'] = (df['Fecha de Egreso'] - df['Fecha de Ingreso']).dt.days


```

```python
# Análisis de datos solo si df está definido
if df is not None:
  # Verificar si la columna 'Tipo de Egreso' existe antes de proceder
  if 'Tipo de Egreso' in df.columns:
    # Distribución por Tipo de Egreso
    tipo_egreso_counts = df['Tipo de Egreso'].value_counts()
    print("\nDistribución por Tipo de Egreso:\n", tipo_egreso_counts)
```

```python
# Graficar la Distribución por Tipo de Egreso
plt.figure(figsize=(10, 6))

plt.title('Distribución de museos por Tipo')
plt.xlabel('Tipo de Egreso')
plt.ylabel('Número de museos')
plt.show()
```

```python
# Graficar Promedio de Duración de Estancia por Departamento
plt.figure(figsize=(10, 6))
plt.title('Promedio de musesos de obras de arte ')
plt.xlabel('Departamento')
plt.ylabel('Duración de Estancia Promedio (días)')
plt.show()
```

```python
# Graficar Número de museos de arte en el pais
plt.figure(figsize=(10, 6))
plt.title('Número de museos por mes de arte en el pais ')
plt.xlabel('Mes en el que ahi mas museos')
plt.ylabel('Número de museos')
plt.show()
```
