```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Intentar leer el archivo CSV especificando la codificación
try:
    df = pd.read_csv('hoja1.csv', encoding='utf-8')
except UnicodeDecodeError:
    # Intentar con una codificación alternativa
    df = pd.read_csv('hoja1.csv', encoding='ISO-8859-1')
# Verificar las primeras filas para asegurarse de que se ha leído correctamente
df.head()

df.head()

# Información general
df.info()

# Estadísticas descriptivas
df.describe()

# Nombres de las columnas
df.columns

# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# Renombrar las columnas para facilitar el análisis
df.columns = ['Año', 'Natalidad']

# Convertir la columna 'Año' a tipo numérico si es necesario
df['Año'] = pd.to_numeric(df['Año'], errors='coerce')

# Convertir la columna 'Natalidad' a tipo numérico si es necesario
df['Natalidad'] = pd.to_numeric(df['Natalidad'], errors='coerce')

# Verificar los tipos de datos después de la conversión
df.dtypes

# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# Gráfico de natalidad por año
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Año', y='Natalidad', marker='o')
plt.title('Tendencia de Natalidad a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Número de Nacimientos')
plt.show()
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Año', y='Natalidad', marker='o')
plt.title('Tendencia de Natalidad a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Número de Nacimientos')
plt.xticks(rotation=45)  # Rote las etiquetas del eje x si es necesario
plt.show()


```
