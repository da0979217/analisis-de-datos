```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


try:
    df = pd.read_csv('/content/hoja1.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('/content/hoja1.csv', encoding='latin-1')


    # Ver las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())

# Ver las últimas filas del DataFrame
print("\nÚltimas filas del DataFrame:")
print(df.tail())


# Obtener un resumen estadístico de las columnas numéricas
print("\nResumen estadístico:")
print(df.describe())


# Obtener información sobre el DataFrame
print("\nInformación del DataFrame:")
print(df.info())


# Ver las columnas disponibles
print("\nColumnas del DataFrame:")
print(df.columns)

# Eliminar filas con valores nulos
df = df.dropna()

# Llenar valores nulos con un valor específico
df = df.fillna(0)

# Eliminar duplicados
df = df.drop_duplicates()

# Cambiar el nombre de una columna
df = df.rename(columns={'nombre_antiguo': 'nombre_nuevo'})


# Histograma
plt.figure(figsize=(10, 6))

plt.title('Histograma de columna')
plt.xlabel('columna')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de dispersión
plt.figure(figsize=(10, 6))

plt.title('Gráfico de dispersión')
plt.xlabel('columna1')
plt.ylabel('columna2')
plt.show()

# Gráfico de caja
plt.figure(figsize=(10, 6))

plt.title('Gráfico de caja de columna')
plt.show()

# Guardar el DataFrame resultante en un nuevo archivo CSV
df.to_csv('datos_limpios.csv', index=False)





```
