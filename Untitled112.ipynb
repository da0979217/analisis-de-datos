```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Configurar el estilo de las gráficas
sns.set(style="whitegrid")

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('egresos hospitalarios.csv', encoding='ISO-8859-1')

# Ver las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())











```

```python
# Ver la información general del DataFrame
print("\nInformación general del DataFrame:")
print(df.info())
```

```python
# Comprobar si hay valores faltantes
print("\nValores faltantes en cada columna:")
print(df.isnull().sum())
```

```python
# Eliminar filas con valores faltantes (si es necesario)
df_cleaned = df.dropna()
```

```python
# Eliminar duplicados
df_cleaned = df_cleaned.drop_duplicates()
```

```python
print("\nDatos después de limpiar:")
print(df_cleaned.info())
```

```python
# Verificar los nombres de las columnas en el DataFrame
print(df_cleaned.columns)
```

```python
# Asumiendo que el nombre correcto de la columna es 'population'
plt.figure(figsize=(12, 6))
```

```python
plt.title('Distribución de egresados del hospital')
plt.xlabel('egresos')
plt.ylabel('egresos hospitalarios')
plt.show()
```

```python
# Asegúrate de usar el nombre correcto para la columna de egresos y los hospitalarios
plt.figure(figsize=(12, 8))
plt.title('Top 10 egresos hospitalarios')
plt.xlabel('egresados')
plt.ylabel('hospitalarios')
plt.show()
```

```python
# Asegúrate de usar el nombre correcto para las columnas de egresados y hospitales
plt.figure(figsize=(12, 8))
plt.title('Relación entre  egresados y hospitalarios')
plt.xlabel('PIB hospitales')
plt.ylabel('hospitales')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
