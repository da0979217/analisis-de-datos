```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# Configurar estilo de los gráficos
sns.set(style="whitegrid")





# Cargar los datos correctamente
df = pd.read_csv('mortalidad.csv', encoding='latin-1')

# Verificar nombres de las columnas
print("Nombres de las columnas en el DataFrame:")
print(df.columns)
```

```python
# Eliminar espacios en blanco en los nombres de las columnas
df.columns = df.columns.str.strip()
```

```python
# Verificar nuevamente los nombres de las columnas
print("Nombres de las columnas después de eliminar espacios en blanco:")
print(df.columns)
```

```python
# Comprobar si las columnas 'demografía' y 'edad' existen en el DataFrame
if 'demografía' in df.columns and 'edad' in df.columns:
  print("Las columnas 'demografía' y 'edad' existen en el DataFrame.")
else:
  print("Error: Las columnas 'demografía' y/o 'edad' no existen en el DataFrame.")
  print("Columnas disponibles en el DataFrame:", df.columns)
```

```python
# Información básica del DataFrame
print(df.info())
```

```python
# Estadísticas descriptivas
print(df.describe())
```

```python
# Verificar valores nulos
print(df.isnull().sum())
```

```python
# Eliminar filas con valores nulos
df_cleaned = df.dropna()
```

```python
# Verificar duplicados
print(f"Número de filas duplicadas: {df_cleaned.duplicated().sum()}")
```

```python
# Eliminar filas duplicadas
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Continuar con el análisis solo si las columnas existen
if 'demografía' in df_cleaned.columns and 'edad' in df_cleaned.columns:
  # Análisis de la distribución por demografía
  plt.figure(figsize=(10, 6))
  sns.countplot(data=df_cleaned, x='demografía', palette='viridis')
  plt.title('Distribución por Demografía')
  plt.xlabel('Demografía')
  plt.ylabel('Frecuencia')
  plt.show()
```

```python
# Análisis de las tasas de mortalidad por periodo
plt.figure(figsize=(12, 8))

plt.title('Tasas de Mortalidad por Periodo')
plt.xlabel('Periodo')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relación entre edad y demografía
plt.figure(figsize=(10, 6))

plt.title('Distribución de Edades por Demografía')
plt.xlabel('Demografía')
plt.ylabel('Edad')
plt.show()
```

```python
# Análisis de la tasa de mortalidad por edad y periodo
plt.figure(figsize=(12, 8))

plt.title('Tasa de Mortalidad por Edad y Periodo')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Correlación entre variables numéricas
plt.figure(figsize=(10, 6))


plt.title('Matriz de Correlación')
plt.show()
```

```python
# Guardar el DataFrame limpio en un nuevo archivo CSV
df_cleaned.to_csv('mortalidad_limpia.csv', index=False)



```
