```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de las gráficas
sns.set(style="whitegrid")


# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('/content/poblacion.csv', encoding='ISO-8859-1')


# Ver las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())

# Ver la información general del DataFrame
print("\nInformación general del DataFrame:")
print(df.info())

# Comprobar si hay valores faltantes
print("\nValores faltantes en cada columna:")
print(df.isnull().sum())

# Eliminar filas con valores faltantes (si es necesario)
df_cleaned = df.dropna()

# Eliminar duplicados
df_cleaned = df_cleaned.drop_duplicates()

print("\nDatos después de limpiar:")
print(df_cleaned.info())

# Verificar los nombres de las columnas en el DataFrame
print(df_cleaned.columns)
# Asumiendo que el nombre correcto de la columna es 'population'
plt.figure(figsize=(12, 6))

plt.title('Distribución de la Población por País')
plt.xlabel('Población')
plt.ylabel('Número de Países')
plt.show()

# Asegúrate de usar el nombre correcto para la columna de población y la de país


plt.figure(figsize=(12, 8))

plt.title('Top 10 Países Más Populosos')
plt.xlabel('Población')
plt.ylabel('País')
plt.show()

# Asegúrate de usar el nombre correcto para las columnas de PIB y población
plt.figure(figsize=(12, 8))

plt.title('Relación entre PIB y Población')
plt.xlabel('PIB')
plt.ylabel('Población')
plt.xscale('log')  # Usar escala logarítmica para el PIB si los valores son muy variados
plt.yscale('log')  # Usar escala logarítmica para la población si los valores son muy variados
plt.show()



```
